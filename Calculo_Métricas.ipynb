```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import roc_curve, auc
```

```python
vp = 180
fp = 30
vn = 130
fn = 40


data = np.array([[vp, fn], [fp, vn]])

plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt="d", cmap="Reds",
            xticklabels=["Positivo", "Negativo"],
            yticklabels=["Verdadeiro/Falso", "Falso/Verdadeiro"],
            cbar=False)

plt.title("Matriz de Confus√£o", fontsize=16)
plt.xlabel("Valor Predito", fontsize=12)
plt.ylabel("Valor Real", fontsize=12)
plt.show()
```

```python
true_labels = [1] * vp + [0] * fn
predicted_scores = np.random.uniform(0, 1, len(true_labels))

fpr, tpr, thresholds = roc_curve(true_labels, predicted_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='b', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Curva ROC')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.legend(loc='lower right')
plt.show()
```

```python

sensibilidade = vp / (vp + fn)
especificidade = vn / (vn + fp)
precisao = vp / (vp + fp)
acuracia = (vp + vn) / (vp + vn + fp + fn)
f_score = 2 * (precisao * sensibilidade) / (precisao + sensibilidade)
roc_auc = auc(fpr, tpr)

print(f"Sensibilidade: {sensibilidade:.2f}")
print(f"Especificidade: {especificidade:.2f}")
print(f"Precis√£o: {precisao:.2f}")
print(f"Acur√°cia: {acuracia:.2f}")
print(f"F-Score: {f_score:.2f}")
print(f"AUC (Area Sobre a Curva): {roc_auc:.2f}")
```
